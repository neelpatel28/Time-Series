```python
import numpy as np
import matplotlib.pyplot as plt
from astsadata import *
from statsmodels.tsa.stattools import acf
from statsmodels.api import OLS, add_constant
```

**EXPERIMENT 3 (a): CHICKEN PRICE SERIES**

```python
chicken = astsadata.chicken["value"].values
t = np.arange(len(chicken))
```

```python
# Original series
plt.figure(figsize=(12, 6)) 
plt.plot(chicken)
plt.title("Chicken Price Time Series")
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
plt.stem(acf(chicken, nlags=30))
plt.title("ACF: Original Chicken Series")
plt.show()
```

```python
# Detrending using OLS
X = add_constant(t)
trend = OLS(chicken, X).fit().predict(X)
detrended = chicken - trend

plt.figure(figsize=(12, 6)) 
plt.stem(acf(detrended, nlags=30))
plt.title("ACF: Detrended Chicken Series")
plt.show()
```

```python
# First difference
diff_chicken = np.diff(chicken)

plt.figure(figsize=(12, 6)) 
plt.stem(acf(diff_chicken, nlags=30))
plt.title("ACF: First Differenced Chicken Series")
plt.show()

print("Chicken Mean (Original):", chicken.mean())
print("Chicken Mean (Detrended):", detrended.mean())
print("Chicken Mean (First Diff):", diff_chicken.mean())
```

**EXPERIMENT 3 (c): GLACIAL VARVES SERIES**

```python
varve = astsadata.varve["value"].values
```

```python
# Original series
plt.figure(figsize=(12, 6)) 
plt.plot(varve)
plt.title("Glacial Varves Time Series")
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
plt.stem(acf(varve, nlags=40))
plt.title("ACF: Glacial Varves")
plt.show()
```

```python
# Log transformation
log_varve = np.log(varve)

plt.figure(figsize=(12, 6)) 
plt.plot(log_varve)
plt.title("Log Glacial Varves")
plt.show()
```

```python
plt.figure(figsize=(12, 6)) 
plt.stem(acf(log_varve, nlags=40))
plt.title("ACF: Log Glacial Varves")
plt.show()
```

```python
# First difference after log
diff_log_varve = np.diff(log_varve)

plt.figure(figsize=(12, 6)) 
plt.stem(acf(diff_log_varve, nlags=40))
plt.title("ACF: Differenced Log Varves")
plt.show()
```
